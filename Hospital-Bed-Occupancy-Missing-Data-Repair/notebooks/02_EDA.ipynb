```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from pathlib import Path
ROOT = Path.cwd().parents[0] if Path.cwd().name == 'notebooks' else Path.cwd()
data = pd.read_csv(ROOT / 'data/processed/hospital_census_clean.csv', parse_dates=['Date'])
```

```python
sns.histplot(data['Bed_Occupancy_Rate'], kde=True); plt.show()
```

```python
data.groupby('Department')['Bed_Occupancy_Rate'].mean().sort_values(ascending=False)
```
